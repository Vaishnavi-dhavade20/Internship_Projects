```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Extract - Load the dataset
heart_data = pd.read_csv("/Heart Disease data.csv")  # Load the dataset from CSV file

# Display the dataset
print(heart_data)
```

```python
# Data Cleaning (Transform)
heart_data.drop_duplicates(inplace=True)  # Remove duplicate rows
heart_data.dropna(inplace=True)          # Remove rows with missing values
```

```python
# EDA - Analyzing heart disease rates
heart_disease_counts = heart_data['target'].value_counts()
print("Heart disease counts:")
print(heart_disease_counts)
```

```python
# EDA - Heart disease by gender
heart_disease_by_gender = heart_data.groupby('sex')['target'].mean()
print("Heart disease by gender:")
print(heart_disease_by_gender)
```

```python
# EDA - Heart disease by age
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='age', data=heart_data)
plt.title('Heart Disease by Age')
plt.xlabel('Heart Disease')
plt.ylabel('Age')
plt.show()
```

```python
# Dashboard
plt.figure(figsize=(18, 12))

# Visualization 1: Heart disease distribution
plt.subplot(3, 3, 1)
sns.countplot(x='target', data=heart_data)
plt.title('Heart Disease Distribution')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
```

```python
# Visualization 2: Heart disease by gender
plt.subplot(1, 1, 1)
sns.barplot(x=heart_disease_by_gender.index, y=heart_disease_by_gender.values)
plt.title('Heart Disease by Gender')
plt.xlabel('Gender')
plt.ylabel('Heart Disease Rate')
```

```python
# Visualization 3: Distribution of Age
plt.subplot(1, 1, 1)
sns.histplot(x='age', data=heart_data, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
# Visualization 4: Distribution of Cholesterol levels
plt.subplot(1, 1, 1)
sns.histplot(x='chol', data=heart_data, kde=True)
plt.title('Distribution of Cholesterol Levels')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
```

```python
# Visualization 5: Resting Blood Pressure (trestbps) vs. Cholesterol (chol)
plt.subplot(1, 1, 1)
sns.scatterplot(x='trestbps', y='chol', data=heart_data, hue='target')
plt.title('Resting BP vs. Cholesterol')
plt.xlabel('Resting BP')
plt.ylabel('Cholesterol')
```

```python
# Visualization 6: Chest Pain Type (cp) Distribution
plt.subplot(1, 1, 1)
sns.countplot(x='cp', data=heart_data, hue='target')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
```

```python
# Visualization 7: Heart Rate (thalach) Distribution
plt.subplot(1, 1, 1)
sns.histplot(x='thalach', data=heart_data, kde=True, hue='target')
plt.title('Heart Rate Distribution')
plt.xlabel('Heart Rate')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
