```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
# Load the dataset
df = pd.read_csv("/content/Financial Analytics data.csv")
```

```python
df
```

```python
# Display the first few rows of the dataframe

print("Initial shape of the dataframe:", df.shape)
print("Initial data types:\n", df.dtypes)
print("\nFirst few rows of the dataframe:")
print(df.head())
```

```python
duplicate_rows = df[df.duplicated()]  #duplicate rows removal
print("\nDuplicate rows:")
print(duplicate_rows)
df
```

```python
# Remove white spaces from column names
df.columns = df.columns.str.strip()
df
```

```python
print("\nSummary statistics:")
print(df.describe())
```

```python
# Convert columns to appropriate data types
df['Mar Cap - Crore']  = pd.to_numeric(df['Mar Cap - Crore'], errors='coerce')
df['Sales Qtr - Crore'] = pd.to_numeric(df['Sales Qtr - Crore'], errors='coerce')
```

```python

# Histogram for Market Capitalization
plt.figure(figsize=(10, 6))
plt.hist(df['Mar Cap - Crore'], bins=20, color='skyblue')
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram for Quarterly Sales
plt.figure(figsize=(10, 6))
plt.hist(df['Sales Qtr - Crore'], bins=20, color='salmon')
plt.title('Distribution of Quarterly Sales')
plt.xlabel('Sales (Crore)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Correlation analysis
correlation = df['Mar Cap - Crore'].corr(df['Sales Qtr - Crore'])
print("Correlation between Market Capitalization and Quarterly Sales:", correlation)
```

```python
# Top performers based on Market Capitalization
top_market_cap_companies = df.nlargest(10, 'Mar Cap - Crore')
print("\nTop Companies based on Market Capitalization:")
print(top_market_cap_companies[['Name', 'Mar Cap - Crore']])

# Top performers based on Quarterly Sales
top_sales_companies = df.nlargest(10, 'Sales Qtr - Crore')
print("\nTop Companies based on Quarterly Sales:")
print(top_sales_companies[['Name', 'Sales Qtr - Crore']])
```

```python
# Bar chart for Top Companies based on Market Capitalization
plt.figure(figsize=(12, 6))
plt.bar(top_market_cap_companies['Name'], top_market_cap_companies['Mar Cap - Crore'], color='orange')
plt.title('Top Companies based on Market Capitalization')
plt.xlabel('Company')
plt.ylabel('Market Capitalization (Crore)')
plt.xticks(rotation=45, ha='right')
plt.show()

# Bar chart for Top Companies based on Quarterly Sales
plt.figure(figsize=(12, 6))
plt.bar(top_sales_companies['Name'], top_sales_companies['Sales Qtr - Crore'], color='green')
plt.title('Top Companies based on Quarterly Sales')
plt.xlabel('Company')
plt.ylabel('Quarterly Sales (Crore)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate key metrics for market capitalization
mean_market_cap = top_market_cap_companies['Mar Cap - Crore'].mean()
median_market_cap = top_market_cap_companies['Mar Cap - Crore'].median()
max_market_cap = top_market_cap_companies['Mar Cap - Crore'].max()
min_market_cap = top_market_cap_companies['Mar Cap - Crore'].min()

# Calculate key metrics for quarterly sales
mean_sales_qtr = df['Sales Qtr - Crore'].mean()
median_sales_qtr = df['Sales Qtr - Crore'].median()
max_sales_qtr = df['Sales Qtr - Crore'].max()
min_sales_qtr = df['Sales Qtr - Crore'].min()

# Display key metrics
print("Key Metrics:")
print("Mean Market Capitalization: {:.2f} Crore".format(mean_market_cap))
print("Median Market Capitalization: {:.2f} Crore".format(median_market_cap))
print("Maximum Market Capitalization: {:.2f} Crore".format(max_market_cap))
print("Minimum Market Capitalization: {:.2f} Crore".format(min_market_cap))
print("Mean Quarterly Sales: {:.2f} Crore".format(mean_sales_qtr))
print("Median Quarterly Sales: {:.2f} Crore".format(median_sales_qtr))
print("Maximum Quarterly Sales: {:.2f} Crore".format(max_sales_qtr))
print("Minimum Quarterly Sales: {:.2f} Crore".format(min_sales_qtr))
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Group by name and calculate average market capitalization and sales
Name = df.groupby('Name').agg({'Mar Cap - Crore': 'mean', 'Sales Qtr - Crore': 'mean'})
print(Name)
```
